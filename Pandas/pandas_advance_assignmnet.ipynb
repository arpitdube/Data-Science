Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column
'Word_Count' that contains the number of words in each row of the 'Text' column.

```python
df = pd.DataFrame({
    "Text" : ["Lorem Ipsum is simply dummy text of the printing and typesetting industry. Lorem Ipsum has been the industry's standard dummy text ever since the 1500s, when an unknown printer took a galley of type and scrambled it to make a type specimen book.",
              "Python is a beautiful programming language",
              "I love Data Science",
              "Things which are equal to the same thing are equal to one another."]
})
```

```python
df
```

```python
def add_col(dataFrame):
    l = []
    l1 = list(dataFrame['Text'])
    for i in l1:
        l.append(len(i.split(' ')))
        
    s = pd.Series(l)
    dataFrame['Word_Count'] = s
```

```python
add_col(df)
```

Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email
addresses in the format 'username@domain.com'. Write a Python function that creates a new column
'Username' in df that contains only the username part of each email address.

The username is the part of the email address that appears before the '@' symbol. For example, if the
email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your
function should extract the username from each email address and store it in the new 'Username'
column.

```python
df = pd.DataFrame({
	"Email": ["amet@outlook.ca","consequat@icloud.ca","libero@icloud.couk","quisque@google.org","elit@outlook.couk","john.doe@example.com"]
})
```

```python
df
```

```python
df['Email'][0].split('@')
```

```python
def extract_username(dataFrame):
    l = []
    for i in dataFrame['Email']:
        l.append(i.split('@')[0])
        
    s = pd.Series(l)
    dataFrame['Username'] = s
```

Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to
create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days
for each row in the DataFrame. The moving average should be calculated using a window of size 7 and
should include the current day.

```python
data = pd.DataFrame({
    'Date': ['2023-08-01', '2023-08-02', '2023-08-04', '2023-08-03', '2023-08-05', '2023-08-06', '2023-08-07', '2023-08-08', '2023-08-09', '2023-08-10', '2023-08-11'],
    'Sales': [100, 150, 200, 120, 180, 220, 130,140,120,200,167]
})
```

```python
def moving_average(df, window_size=7):
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.sort_values(by='Date')
    df['MovingAverage'] = df['Sales'].rolling(window=window_size, min_periods=1).mean()
    return df
```

```python
moving_average(data)
```

```python
dates = pd.date_range(start = '2023-01-01', end = '2023-01-08')

df = pd.DataFrame({'Date' : dates})
```

```python
def weekDay(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df['Weekday'] = df['Date'].dt.day_name()
    return df
```

Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python
function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

```python
data = pd.DataFrame({'Date': ['2023-01-15 08:00:00', '2023-01-25 12:30:00', '2023-02-05 15:45:00', '2023-01-10 10:15:00', '2023-02-01 14:00:00']})
```

```python
start_timestamp_date = '2023-01-01 00:00:00'
end_timestamp_date = '2023-01-31 23:59:59'
```

```python
def select_date(df,start_timestamp_date,end_timestamp_date):
    df['Date'] = pd.to_datetime(df['Date'])
    start_timestamp_date = pd.to_datetime(start_timestamp_date)
    end_timestamp_date = pd.to_datetime(end_timestamp_date)
    answer = df[(df['Date'] >= start_timestamp_date) & (df['Date'] <= end_timestamp_date)]
    return answer
```

```python
select_date(data,start_timestamp_date,end_timestamp_date)
```
